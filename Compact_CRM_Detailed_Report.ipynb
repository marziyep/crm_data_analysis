```python
net_date=pd.DataFrame(crm.groupby("Date")["Net"].sum())
```

```python
cost_date=pd.DataFrame(cost.groupby('Date')['Cost'].sum())
```

```python
net_cost_date=net_date.merge(cost_date,how="left",left_index=True,right_index=True)
```

```python
net_cost_date.plot()
plt.ylabel("sum_of_money")
plt.plot()
```

```python
net_cost_date
```

```python
net_cost_date["conversation_rate"]=(net_cost_date["Net"]/net_cost_date["Cost"])*100
```

```python
net_cost_date["conversation_rate"].plot(kind="bar",color="Green")
plt.plot()
```

```python
net_system=pd.DataFrame(crm.groupby("System")["Net"].sum())
```

```python
cost_system=pd.DataFrame(cost.groupby("System")["Cost"].sum())
```

```python
system_net_cost=net_system.merge(cost_system,how="left",left_index=True,right_index=True)
```

```python
system_net_cost.fillna(0,inplace=True)
```

```python
system_net_cost.plot(kind="bar",color=["brown","blue"])
plt.ylabel("sum_of_money")
plt.plot()
```

```python
system_net_cost.head()
```

```python
a=[]
for i in system_net_cost.index:
    if system_net_cost.loc[i,"Cost"]==0 :
       a.append (system_net_cost.loc[i,"Net"])
    else:
        a.append(system_net_cost.loc[i,"Net"]/system_net_cost.loc[i,"Cost"])
      
 
```

```python
system_net_cost["conversion_rate"]=a
```

```python
system_net_cost["conversion_rate"].plot(kind="bar")
plt.plot()
```
